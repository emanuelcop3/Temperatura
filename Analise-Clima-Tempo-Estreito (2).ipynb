```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# Localizamos uma cidade usando o próprio serviço de clima-tempo
url = "https://www.tempo.com/estreito_maranhao-l108203.htm"
```

```python
# usamos a biblioteca requests para fazer o download da página web
page = requests.get( url )
page
```

```python
# criamos um objeto beautiful soup para poder fazer a raspagem dos dados
soup = BeautifulSoup(page.content, 'html.parser')
#soup
print(soup.prettify())
```

```python
# após investigar, com um navegador web, a estrutura da página e localizar o elemento HTML que queremos, aplicamos o comando
datos_dos = soup.find(class_="datos-dos-semanas")
datos_dos
```

## Extraindo toda a informação da página

```python
datos_dos
```

```python
datos_dos.select(".cuando")
```

```python
# Usa o seletor CSS para extrair todos os period-name dentro do cuando
cuando_tags = datos_dos.select(".cuando")

# Usando list comprehensions (não vimos ainda)
#periods = [pt.get_text() for pt in period_tags]

cuando = []
for pt in cuando_tags:
    cuando.append(pt.get_text())
cuando
```

```python
# Usando list comprehensions (não vimos ainda)
#short_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]

temperatura_tags = datos_dos.select(".temperatura")
temperatura = []
for sd in temperatura_tags:
    temperatura.append(sd.get_text())
temperatura
```

```python
# Usando list comprehensions (não vimos ainda)
# temps = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]

changeUnitW_tags = datos_dos.select(".changeUnitW")
changeUnitW = [] 
for tp in changeUnitW_tags:
    changeUnitW.append(tp.get_text())
changeUnitW
```

## Média de temperatura em Estreito no Maranhão

```python
df_clima_tempo_Estreito["Temperatura"]
```

```python
# Neste caso, a expressão regular para extrair os digitos da string 
temperatura_nums = df_clima_tempo_Estreito["Temperatura"].str.extract(r"(\d+)", expand=False)
temperatura_nums
```

```python
# O tipo resultante de Series.str.extract
type(temperatura_nums)
```

```python
# Converte os valores da coluna temp_num em inteiro
df_clima_tempo_Estreito["Temperatura"] = temperatura_nums.astype('int')
df_clima_tempo_Estreito["Temperatura"]
```
